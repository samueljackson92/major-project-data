```python
%matplotlib qt
import pandas as pd
import numpy as np
import mia
```

```python
b1 = pd.DataFrame.from_csv("/Volumes/Seagate/results/2015-03-29/2015-03-29-real-blobs1.csv")
b2 = pd.DataFrame.from_csv("/Volumes/Seagate/results/2015-03-29/2015-03-29-real-blobs2.csv")
hologic = pd.concat([b1, b2])
hologic.head()
```

```python
hologic_meta = mia.analysis.create_hologic_meta_data(hologic, "/Volumes/Seagate/meta_data/BIRADS.csv")
hologic_meta.head()
```

```python
hologic_intensity = hologic[hologic.columns[3:]]
hologic.loc['p214-010-60001-cl.png']
```

## Group intensity by just the image

Taking the mean of all intensity features, across all scales.

```python
group = hologic_intensity.groupby(hologic_meta.index)
features = group.apply(lambda x: x.mean())
features.head()
```

```python
mapping = mia.analysis.tSNE(features, verbose=2, learning_rate=400, perplexity=30)
```

```python
mia.plotting.plot_scatter_2d(mapping, [0,1], hologic_meta.drop_duplicates().BIRADS)
```

```python
hologic.dropna(inplace=True)
group = hologic.groupby([hologic.radius, hologic.index])
g = group.apply(lambda x: x.dropna().mean())
g
```
